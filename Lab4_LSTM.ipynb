```python
#Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

print("TensorFlow version:", tf.__version__)
```

```python
corpus = [
"machine learning is the future of technology",
"deep learning helps in image recognition",
"natural language processing is very useful",
"artificial intelligence is transforming industries",
"data science is important for decision making",
"python is widely used in machine learning",
"neural networks learn patterns from data",
"lstm is used for sequence prediction",
"rnn works well with text data",
"ai is used in healthcare and finance",
"big data helps companies grow",
"technology is changing the world",
"students are learning artificial intelligence",
"cloud computing supports big data",
"deep learning models need more data",
"tensorflow is a popular framework",
"pytorch is also widely used",
"data analysis requires clean data",
"machine learning models need training",
"algorithms learn from examples",
"text generation is done using lstm",
"sequence models predict next word",
"ai is the future of innovation",
"automation is increasing in industries",
"robots are powered by ai",
"data engineers manage pipelines",
"data analysts create reports",
"neural networks have many layers",
"deep learning improves accuracy",
"training data is very important",
"prediction models require testing",
"lstm remembers past information",
"rnn has memory capability",
"artificial intelligence helps humans",
"technology makes life easier",
"students study data science",
"programming is essential for ai",
"python is easy to learn",
"learning never stops in technology",
"innovation drives the future"
]

print("Number of sentences:", len(corpus))
print("\nSample corpus:")
for line in corpus:
    print("-", line)
```

```python
#Tokenization
#it completes the setences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)

word_index = tokenizer.word_index
total_words = len(word_index) + 1  # +1 because indexing starts from 1

print("Vocabulary size:", total_words)
print("\nWord index:")
print(word_index)
```

```python
input_sequences = []

for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

print("Total training sequences:", len(input_sequences))
print("\nSome sequences before padding:")
for seq in input_sequences[:10]:
    print(seq)
```

```python
max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

print("Maximum sequence length:", max_seq_len)
print("\nPadded sequences:")
print(input_sequences[:10])
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = tf.keras.utils.to_categorical(y, num_classes=total_words)

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
model = Sequential([
    Embedding(input_dim=total_words, output_dim=10, input_length=max_seq_len - 1),
    LSTM(100),
    Dense(total_words, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X, y, epochs=300, verbose=1)
```

```python
print("Final Accuracy:", history.history['accuracy'][-1])
```

```python
def generate_text(seed_text, next_words):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len - 1, padding='pre')

        predicted_probs = model.predict(token_list, verbose=0)
        predicted_index = np.argmax(predicted_probs, axis=-1)[0]

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break

        seed_text += " " + output_word
    return seed_text
```

```python
print(generate_text("deep learning", 2))
print(generate_text("machine learning", 2))
print(generate_text("python is", 2))
print(generate_text("artificial intelligence", 2))
print(generate_text("data science", 2))
print(generate_text("neural networks", 2))
```
